# Data wrangling of the new miRNA features
In this step, the individual sets with the new miRNAs features associated with E-cadherin deregulation and EMT will be concatenated. The result of this combination will be merged with the existing df_mir_expression to include both the original and new features in a single dataset ready for modeling.

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
import pandas as pd
```

```python
# Reading stage I to II df
df_stages_I_II_new = pd.read_csv('df_stages_I_II_new.csv', index_col='File_ID')
df_stages_I_II_new.head()
```

```python
# Reading stage III to IV
df_stages_III_IV_new = pd.read_csv('df_stages_III_IV_new.csv', index_col='File_ID')
df_stages_III_IV_new.head()
```

```python
# Concatenating both stages dfs
df_mir_expression_new = pd.concat([df_stages_I_II_new, df_stages_III_IV_new], axis=0)
df_mir_expression_new.head()
```

```python
# Checking appropiate merging
df_mir_expression_new.shape
```

```python
# Reading wrangled data frame with the base miRNA features
df_mir_expression = pd.read_csv('df_mir_expression_wrangled.csv', index_col='File_ID')
df_mir_expression.head()
```

```python
# Check for correct number of rows and columns
df_mir_expression.shape
```

```python
#Check for and handle duplicate index values in df_mir_expression
if not df_mir_expression.index.is_unique:
    print("Warning: df_mir_expression has duplicate index values. Dropping duplicates.")
    df_mir_expression = df_mir_expression.loc[~df_mir_expression.index.duplicated(keep='first')]

# Check for and handle duplicate index values in df_mir_expression_new
if not df_mir_expression_new.index.is_unique:
    print("Warning: df_mir_expression_new has duplicate index values. Dropping duplicates.")
    df_mir_expression_new = df_mir_expression_new.loc[~df_mir_expression_new.index.duplicated(keep='first')]

# Now concatenate the dataframes
df_mir_new = pd.concat([df_mir_expression, df_mir_expression_new], axis=1)
```

```python
# Ensure proper concatenation of the new features along axis 1
df_mir_new.shape
```

```python
# Set the columns that need to be rounded
round_columns = df_mir_new.columns
round_columns = round_columns[:24]
# Build the for loop rounding each miRNA expression column
for column in round_columns:
  df_mir_new[column] = np.round(df_mir_new[column], decimals=3)
```
